## 2. Clone the ROOT repository and build it

These processes take approximately **4 hours**.

```python
%cd /opt
!git clone --progress https://github.com/root-project/root.git -b v6-24-06 --depth 1 root-src
%rm -rd root-src/.git
```

```python
%mkdir /opt/root_build /opt/root
%cd /opt/root_build
!cmake -Dpython3=ON \
    -Dimt=OFF \
    -Dbuiltin_tbb=OFF \
    -Dmathmore=OFF \
    -DPYTHON_EXECUTABLE:FILEPATH=/usr/bin/python3 \
    -DPYTHON_INCLUDE_DIR:PATH=/usr/include/python3.7m \
    -DPYTHON_INCLUDE_DIR2:PATH=/usr/include/x86_64-linux-gnu/python3.7m \
    -DPYTHON_LIBRARY:FILEPATH=/usr/lib/x86_64-linux-gnu/libpython3.7m.so \
    -DCMAKE_INSTALL_PREFIX=/opt/root \
    /opt/root-src
```

```python
%%time
%cd /opt/root_build
!cmake --build . -- install
```

```python
with tarfile.open("root-bin_colab.tar.xz", 'x:xz') as t:
    t.add("/opt/root", arcname="root")
with tarfile.open("root-src_colab.tar.xz", 'x:xz') as t:
    t.add("/opt/root-src", arcname="root-src")
```
